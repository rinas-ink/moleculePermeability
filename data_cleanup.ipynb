```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
molecules = pd.read_csv('molecules_dataset.csv')
```

```python
print(molecules.shape)
print(molecules.dtypes)
molecules.columns = ['id'] + molecules.columns[1:].tolist()
```

```python
molecules.sample(3)
```

```python
molecules.nunique() == molecules.shape[0]
```

```python
def show_duplicates(data):
    grouped = data.groupby(by='SMILES').nunique()
    is_unique = grouped == 1
    strange_rows = is_unique.loc[~is_unique.all(axis=1)].index
    strange_cols = is_unique.loc[strange_rows, ~is_unique.all(axis=0)].columns
    strange_cols = ['SMILES'] + strange_cols.tolist()
    strange_data = molecules.copy()
    strange_data = strange_data.loc[strange_data['SMILES'].isin(strange_rows), strange_cols].sort_values(by='SMILES')
    strange_data['Group'] = strange_data.groupby(by='SMILES').ngroup()
    return strange_data

show_duplicates(molecules)
```

```python
cleared_molecules = molecules.drop(columns=['img_file'])
cols = [col for col in cleared_molecules.columns if col not in ['id']]
cleared_molecules.drop_duplicates(inplace=True, subset=cols)
show_duplicates(cleared_molecules)
```

```python
completely_cleared = cleared_molecules.drop([28, 55, 5, 23, 135])
show_duplicates(completely_cleared)
```

```python
completely_cleared.to_csv('molecules_dataset_cleared.csv', index=False)
```
